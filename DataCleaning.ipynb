```python
import numpy as np
import pandas as pd

data = pd.read_csv('/Users/jay/Desktop/Spring 2023/ML/My projects/glassdoor_jobs.csv')
data.columns
# data.describe()
```

## Salary Estimate Column Cleaning

```python
data = data[data['Salary Estimate'] != '-1']
data
```

```python
# removing the (Glassdoor est) from the Salary Estimate Column

salary = data['Salary Estimate'].apply(lambda x: x.split('(')[0])

# removing k and $ signs and replacing them with space

rmv_k = salary.apply(lambda x: x.replace('$','').replace('K','')) 

# salary
```

```python
data['hourly'] = data['Salary Estimate'].apply(lambda x: 1 if 'per hour' in x.lower() else 0)
data['employer_provided'] = data['Salary Estimate'].apply(lambda x: 1 if 'employer provided' in x.lower() else 0)
data
```

```python
# removing per hour and employee provided from Salary estimate
rmv_hr = rmv_k.apply(lambda x: x.lower().replace('per hour','').replace('employer provided salary:',''))
data['min_salary'] = rmv_hr.apply(lambda x: int(x.split('-')[0]))
data['max_salary'] = rmv_hr.apply(lambda x: int(x.split('-')[1]))
data['avg_salary'] = (data.min_salary + data.max_salary)/2
data
```

## Company Name Cleaning

```python
data['comapany_txt'] = data.apply(lambda x: x['Company Name'] if x['Rating'] < 0 else x['Company Name'][:-3],axis=1)


                                                            
```

## Get state from the location column

```python
data['state'] = data['Location'].apply(lambda x: x.split(',')[1])
data.state.value_counts()
# checkinf if the headquarters and job location are in the same state

data['same_state'] = data.apply(lambda x : 1 if x['state'] == x['Headquarters'] else 0,axis=1)
data
```

```python
# age of the company

data['age'] = data['Founded'].apply(lambda x: x if x < 1 else 2023 - x )
data
```
